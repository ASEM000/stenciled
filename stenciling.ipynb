## a) Basic operations

### 1) Identity

```python
@stenciled()
def identitiy(x):
    return x[0,0]

X = np.arange(1,26).reshape(5,5)
print(f'X = \n {X} \n\n identity(X) = \n {identitiy(X)}')
```

### 2) Mean filter

```python
@stenciled()
def mean(x):
    return ( x[1,-1] + x[1,0] + x[1,1] +
             x[0,-1] + x[0,0] + x[0,1] +
             x[-1,-1] +x[-1,0] + x[-1,1]) / 9

X = np.arange(1,26).reshape(5,5)
print(f'X = \n {X} \n\n mean(X) = \n {mean(X)}')
```

### 3) Laplacian filter

```python
@stenciled()
def laplacian(x):
    return ( 0*x[1,-1]  + 1*x[1,0]   + 0*x[1,1] +
             1*x[0,-1]  +-4*x[0,0]   + 1*x[0,1] +
             0*x[-1,-1] + 1*x[-1,0]  + 0*x[-1,1] ) 

np.random.seed(0)
X = np.random.randint(low=0,high=150,size=(5,5))
print(f'X = \n {X} \n\n laplacian(X) = \n {laplacian(X)}')
```

- **1D Linear convection**
$\large \frac{\partial u}{\partial t} + c \frac{\partial u}{\partial x} = 0 \Longrightarrow u_i^{n+1} = u_i^n - c \frac{\Delta t}{\Delta x}(u_i^n-u_{i-1}^n)$

```python
#Initial U

tmax,xmax=0.5,2
nt,nx= 101,51

# Increments
dt = tmax/(nt-1)
dx = xmax/(nx-1)
c  = 0.5
nu = 0.1

u = np.zeros((nx,nt))
x = np.zeros(nx)

# Boundary conditions
u[0,:] = u[nx-1,:] = 1

# Initial conditions
for i in range(1,nx-1):
    if(i > (nx-1)/4 and i < (nx-1)/2):
        u[i,0] = 2
    else:
        u[i,0] = 1
```

```python
@stenciled(inplace=True,border='all',window=(3,3))
def linear_convection(u,k):
    return u[-1,0] - k * (u[-1,0] - u[-1,-1])
```

```python
plt.figure(figsize=(15,5))
res0 = linear_convection(u.copy().T,c*dt/dx)[:-1]
plt.xlabel('t(s)');plt.ylabel('u(m/s)');plt.title('Linear convection')
for i in range(0,res0.shape[0],20): plt.plot(np.linspace(0,xmax,nx),res0[i])
```

- **1D Nonlinear convection**
$\large \frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = 0 \Longrightarrow u_i^{n} = u_i^{n-1} - u_i^{n-1} \frac{\Delta t}{\Delta x} (u_i^{n-1} - u_{i-1}^{n-1})$

```python
@stenciled(inplace=True,window=(3,3),border='all')
def nonlinear_convection(u,k):return u[-1,0] - u[-1,0]*k * (u[-1,0] - u[-1,-1])
```

```python
plt.figure(figsize=(15,5))
res1 = nonlinear_convection(u.copy().T,dt/dx)[:-1]
plt.xlabel('t(s)');plt.ylabel('u(m/s)');plt.title('Non linear convection')
for i in range(0,res1.shape[0],20): plt.plot(np.linspace(0,xmax,nx),res1[i])
```

- **1D Diffusion**
$\large \frac{\partial u}{\partial t}= \nu \frac{\partial^2 u}{\partial x^2} \Longrightarrow u_{i}^{n}=u_{i}^{n-1}+\frac{\nu\Delta t}{\Delta x^2}(u_{i+1}^{n-1}-2u_{i}^{n-1}+u_{i-1}^{n-1})$ 


```python
@stenciled(inplace=True,window=(3,3),border='all')
def diffusion_1D(u,k): return u[-1,0] + k * (u[-1,1] -2*u[-1,0] + u[-1,-1])
```

```python
plt.figure(figsize=(15,5))
res2 = diffusion_1D(u.copy().T,nu*dt/dx**2)[:-1]
plt.xlabel('t(s)');plt.ylabel('u(m/s)');plt.title('1D Diffusion')
for i in range(0,res2.shape[0],20): plt.plot(np.linspace(0,xmax,nx),res2[i])
```
